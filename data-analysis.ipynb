### Importação de Pacotes Necessários

---

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Armazenando o Dataset em uma Variável

```python
dados = pd.read_csv('dataset/Global_Cybersecurity_Threats_2015-2024.csv')
```

### Análise Exploratória

```python
# Tamanho do Dataframe
dados.shape
print(f'O dataset possui {dados.shape[0]} registros (linhas) e {dados.shape[1]} atributos (colunas)')
```

```python
# Verificando os primeiros registros do dataset
dados.head (10)
```

```python
# Traduzindo o conteúdo do dataset através de um dicionário

traducao_colunas = {
    "Country": "País",
    "Year": "Ano",
    "Attack Type": "Tipo de Ataque",
    "Target Industry": "Indústria Alvo",
    "Financial Loss (in Million $)": "Prejuízo Financeiro (em Milhões de $)",
    "Number of Affected Users": "Número de Usuários Afetados",
    "Attack Source": "Fonte do Ataque",
    "Security Vulnerability Type": "Tipo de Vulnerabilidade de Segurança",
    "Defense Mechanism Used": "Mecanismo de Defesa Utilizado",
    "Incident Resolution Time (in Hours)": "Tempo de Resolução do Incidente (em Horas)"
}
```

```python
# Renomeando as colunas com base no dicionário criado

dados.rename(columns=traducao_colunas, inplace=True)
```

```python
# Verificando novamente os primerios registro do dataset

dados.head(10)
```

```python
# Verificando os tipos de dados de cada coluna
dados.info()
```

```python
# Estatísticas descritivas das colunas com dados numéricos
dados.describe()
```

```python
# Descobrindo o maior prejuízo financeiro e em qual país/ano aconteceu
maior_prejuizo = dados['Prejuízo Financeiro (em Milhões de $)'].max()
dados_completos = dados[dados['Prejuízo Financeiro (em Milhões de $)'] == maior_prejuizo]
print(dados_completos.to_string())
```

```python
# Melhorando a visualização anterior 
print(dados_completos.iloc[0])
```

```python
# Listando os ataques mais comuns
dados['Tipo de Ataque'].value_counts()
```

```python
# Verificando o país que mais aparece nos ataques mais comuns

combinacoes = dados.value_counts(subset=['País', 'Tipo de Ataque'])

combinacoes_df = combinacoes.reset_index(name='Quantidade')
print(combinacoes_df)
```

```python
dados['País'].value_counts()
```

```python
# Representação gráfica das ocorrência por País e tipo de ataque

plt.figure(figsize=(12, 6))
ax = sns.countplot(x='País', hue='Tipo de Ataque', data=dados)

plt.title('Ocorrência por País e Tipo de Ataque')
plt.xlabel('País')
plt.ylabel('Número de Ocorrências')

plt.legend(title='Tipo de Ataque', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()  
plt.show()
```

```python
# Gráfico para representação de números e tipos de ataque
plt.figure(figsize=(14, 7))

sns.countplot(x='Ano', hue='Tipo de Ataque', data=dados)

plt.title('Evolução do Número de Ataques por Ano e Tipo de Ataque')
plt.xlabel('Ano')
plt.ylabel('Número de Ataques')
plt.legend(title='Tipo de Ataque', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
